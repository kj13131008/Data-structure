1. What are data structures, and why are they important?

8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table is a data structure that stores data in key–value pairs using a hash function to determine where each key’s value is stored in memory. The hash function converts a key into a numeric hash, enabling fast data access. Python dictionaries are implemented using hash tables. When a key is added to a dictionary, Python hashes it and stores the value at the corresponding location. This allows dictionaries to support very fast lookup, insertion, and deletion operations, typically with O(1) average time complexity, making them highly efficient for data retrieval.

17. Describe the time complexity of accessing elements in a dictionary.
  - In Python, accessing elements in a dictionary has an average time complexity of O(1), which means constant time. This is because dictionaries use a hash table internally, where keys are converted into hash values to determine their storage location. As a result, retrieving a value using a key is usually very fast, regardless of the dictionary size. However, in rare cases of many hash collisions, the time complexity can degrade to O(n), where n is the number of elements. Python minimizes collisions using efficient hashing and resizing strategies.
18. In what situations are lists preferred over dictionaries?
  - Lists are preferred over dictionaries when data needs to be stored in an ordered sequence and accessed by position or index. They are ideal when elements are processed sequentially, such as iterating through items or applying operations in order. Lists are also suitable when data does not require key-value mapping and when duplicates are allowed. They use less memory compared to dictionaries, making them efficient for simple collections. Additionally, lists are preferred when the dataset is small or when frequent slicing, sorting, or indexing operations are required.
19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Dictionaries are considered unordered because they store data based on hash values, not insertion order. This means elements are accessed by keys rather than position. As a result, data retrieval is fast and efficient, but indexing or positional access is not supported.
20. Explain the difference between a list and a dictionary in terms of data retrieval.
  - A list and a dictionary differ mainly in how data is retrieved. In a list, elements are accessed using their index positions, such as my_list[2], which works well when the position of data is known. However, finding a specific value in a list may require scanning elements one by one, taking O(n) time. In contrast, a dictionary retrieves data using unique keys, like my_dict["name"]. This key-based access uses hashing, allowing much faster lookup, usually in O(1) time, making dictionaries ideal for quick data retrieval.
